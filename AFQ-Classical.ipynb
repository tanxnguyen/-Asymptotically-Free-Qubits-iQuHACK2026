```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import yfinance as yf
from scipy.stats import norm
```

```python
# test asset return pdf
def asset_return_pdf(mu, sigma):
    return norm(loc=mu, scale=sigma)
```

```python
# estimated parameters
mu = 0.15      # ~15% daily return
sigma = 0.2     # ~20% daily volatility
V0 = 1_000    # portfolio value
alpha = 0.95      # confidence level


r = np.linspace(-0.1, 0.1, 500)
pdf = asset_return_pdf(r, mu, sigma)
```

```python
def get_VaR_theory(mu, sigma, V0=1.0, alpha=0.95):
    z_alpha = norm.ppf(1 - alpha)
    VaR_theory = -V0 * (mu + sigma * z_alpha)
    return VaR_theory
```
